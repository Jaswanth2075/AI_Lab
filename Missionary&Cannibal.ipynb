```python
class Miscal:
    def __init__(self, lm, lc, rm, rc, boat):
        self.lm = lm
        self.lc = lc
        self.rm = rm
        self.rc = rc
        self.boat = boat

    def check(self, a, b, c, d):
        return (a >= 0 and b >= 0 and c >= 0 and d >= 0 and (a == 0 or a >= b) and (c == 0 or c >= d))

    def newstate(self, frontier, curr_path, na, nb, nc, nd, new_bt):
        if self.check(na, nb, nc, nd):
            frontier.append((na, nb, nc, nd, new_bt, curr_path))

    def ronem(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a - 1, b, c + 1, d, 1)

    def ronec(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a, b - 1, c, d + 1, 1)

    def rtwom(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a - 2, b, c + 2, d, 1)

    def rtwoc(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a, b - 2, c, d + 2, 1)

    def ronemonec(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a - 1, b - 1, c + 1, d + 1, 1)

    def lonem(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a + 1, b, c - 1, d, 0)

    def lonec(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a, b + 1, c, d - 1, 0)

    def ltwom(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a + 2, b, c - 2, d, 0)

    def ltwoc(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a, b + 2, c, d - 2, 0)

    def lonemonec(self,frontier,curr_path,a,b,c,d,bo):
        self.newstate(frontier, curr_path, a + 1, b + 1, c - 1, d - 1, 0)

    def bfs(self):
        explored = set()
        frontier = [(self.lm, self.lc, self.rm, self.rc, self.boat, [])]

        while frontier:
            a, b, c, d, bt, path = frontier.pop()

            if a == 0 and b == 0 and c == 3 and d == 3 and bt == 1:
                print("Crossed Successfully...\n")
                print(path+[(a,b,c,d,bt)])
                return

            if (a, b, c, d, bt) not in explored and (a,b,c,d,bt) not in frontier:
                explored.add((a, b, c, d, bt))
                curr_path = path + [(a, b, c, d, bt)]

                if bt == 0:
                    self.ronem(frontier, curr_path, a, b, c, d, bt)
                    self.ronec(frontier, curr_path, a, b, c, d, bt)
                    self.rtwom(frontier, curr_path, a, b, c, d, bt)
                    self.rtwoc(frontier, curr_path, a, b, c, d, bt)
                    self.ronemonec(frontier, curr_path, a, b, c, d, bt)

                if bt == 1:
                    self.lonem(frontier, curr_path, a, b, c, d, bt)
                    self.lonec(frontier, curr_path, a, b, c, d, bt)
                    self.ltwom(frontier, curr_path, a, b, c, d, bt)
                    self.ltwoc(frontier, curr_path, a, b, c, d, bt)
                    self.lonemonec(frontier, curr_path, a, b, c, d, bt)


lm = 3
lc = 3
rm = 0
rc = 0
boat = 0
Miscal(lm, lc, rm, rc, boat).bfs()
```
